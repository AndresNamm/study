## VISUALIZATION OF THE CORRELATION FUNCTION 


Correlation = $ \frac{Cov(X,Y)}{SD(X)SD(Y)}$ = $\frac{\sum (X-\overline{X})(Y-\overline{Y})}{\sqrt{\sum (X-\overline{X})^2}\sqrt{\sum (Y-\overline{Y})^2}}$



+ The top and bottom of this function in case of high correlation should be equalt to each other. 
+ In the bottom we are taking the sum of deviations squared and then taking the sqrt from it $\sqrt{\sum (X-\overline{X})^2}\sqrt{\sum (Y-\overline{Y})^2}$.  This is less than just taking the sum of deviations. $\sum |(X-\overline{X})|\sum |(Y-\overline{Y})|$ =>  $\sqrt{\sum (X-\overline{X})^2}\sqrt{\sum (Y-\overline{Y})^2}$ <  $\sum |(X-\overline{X})|\sum |(Y-\overline{Y})|$
+ Interesting observations  
   + $\sqrt{\sum (X-\overline{X})^2} < \sum |(X-\overline{X})|$
   + $\sum (X-\overline{X})(Y-\overline{Y}) < \sum (X-\overline{X}) \sum (Y-\overline{Y})$


```python
def calculateSum(t):
    sum = 0 
    for i in range(t):
        sum += i*i
    return sum
def calculateAddition(t):
    sum1 = 0 
    for i in range(t):
        sum1 += i
    sum2 = 0 

    for i in range(t):
        sum2 += i

    return sum1*sum2
def calculateAdditionSquared(t):
    sum1 = 0 
    for i in range(t):
        sum1 += i*i
    sum2 = 0 

    for i in range(t):
        sum2 += i*i
    return math.sqrt(sum1)*math.sqrt(sum2)

start=0
finish=5
x = np.arange(start, finish, 1)

y = np.array([calculateSum(i) for i in range(start,finish)])
y2 = np.array([calculateAddition(i) for i in range(start,finish)])
y3 = np.array([calculateAdditionSquared(i) for i in range(start,finish)])
overlapping = 0.150
print('Values of x: ', x)
print('Values of y: ', y)
plt.plot(x, y,label = "Corr Top (Replicate and Sum)", linewidth=8)
plt.plot(x, y3,label = "Corr Bottom Squared (Square, Sum, Sqrt and Replicate)",linewidth=5)
plt.plot(x, y2,label = "Corr Bottom (Sum and Replicate)",linewidth=6)

plt.title("Correlation")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.legend()
plt.show()
```

$\sqrt{\sum (X-\overline{X})^2} < \sum |(X-\overline{X})|$  

$\sqrt{\sum (t)^2} < \sum |(t)|$

```python
def calculateAdditionNoMultiplication(t):
    sum1 = 0 
    for i in range(t):
        sum1 += i
    sum2 = 0 
    return sum1

def calculateAdditionSqrdNoMultiplication(t):
    sum1 = 0 
    for i in range(t):
        sum1 += i*i
    sum2 = 0 
    return math.sqrt(sum1)


start=0
finish=100
x = np.arange(start, finish, 1)

y2 = np.array([calculateAdditionNoMultiplication(i) for i in range(start,finish)])
y3 = np.array([calculateAdditionSqrdNoMultiplication(i) for i in range(start,finish)])
overlapping = 0.150
print('Values of x: ', x)

plt.plot(x, y2,label = "Regular Sum",linewidth=6)
plt.plot(x, y3,label = "Squared Sum",linewidth=5)
plt.title("Correlation")
plt.xlabel("Values of x")
plt.ylabel("Values of y")
plt.legend()
plt.show()
```

The previous is because the  derivative of the SQRT function is getting smaller. Aka the growth is slowing down as the input to the SQRT function grows. 

+ As reference derivative for X=X function is 1. The growth happens linearly endlessly.
