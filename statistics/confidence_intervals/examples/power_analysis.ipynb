## POWER ANALYSIS FOR CONFIDENCE INTERVALS 

### CONTINUOUS DATA - 1 SAMPLE 


```python
from scipy import stats

# e = margin of error
# std_dev = standard deviation of the population 
# alpha = significance level of the test (1 - confidence level) 
# beta = power of the test
def get_sample_size(e, std_dev, alpha, beta):
    """
    Returns the sample size needed to achieve a given margin of error
    with a given standard deviation, alpha and beta.
    """
    z_value = stats.norm.ppf(1 - alpha / 2)
    n = (z_value*std_dev/e)**2  
    return n
```

#### EXAMPLE 1

Power analysis for confidence interval with continuous data and 1 sample in Python. The dataset is average height in a country.Here we are showcasing what the sample size needs to be for 
different E values.

```python
# Plot how the sample size changes with the margin of error from  0.1 to 2 assuming ALPHA and KNOWN STD_DEV are constant
import matplotlib.pyplot as plt
import numpy as np

KNOWN_STD_DEV = 2  # This is knownd from previous experiments
ALPHA = 0.05 # 95% confidence interval

x = np.linspace(0.1, 2, 100)
y = [get_sample_size(m, KNOWN_STD_DEV, ALPHA, 0.2) for m in x]


# Set up the plot
plt.plot(x, y)

# Add labels
plt.xlabel('Margin of Error')
plt.ylabel('Sample Size')
plt.title('Sample Size vs Margin of Error for a 95% Confidence Interval')
```

#### EXAMPLE 2

An investigator wants to estimate the mean systolic blood pressure in children with congenital heart disease who are between the ages of 3 and 5. 
How many children should be enrolled in the study? The investigator plans on using a 95% confidence interval (so Z=1.96) and wants a margin of error of 5 units. 
The standard deviation of systolic blood pressure is unknown, but the investigators conduct a literature search and 
find that the standard deviation of systolic blood pressures in children with other cardiac defects is between 15 and 20. To estimate the sample size, we consider the larger standard deviation in order to obtain the most conservative (largest) sample size. 

+ [Source](https://sphweb.bumc.bu.edu/otlt/mph-modules/bs/bs704_power/bs704_power_print.html) by Lisa Sullivan, PhD 

```python
# Calculate the sample size
sample_size = get_sample_size(5, 20, 0.05, 0.2)
print(f'Sample size: {sample_size}')
```
